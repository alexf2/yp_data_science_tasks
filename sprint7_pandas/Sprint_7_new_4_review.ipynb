 ### Общие функции

````python
def show_ds_short_info(df):
    """
        Выводит информацию о размерности датасета и перечень колонок.
    """
    print(
        f'\tВсего: {df.shape[0]} строк, {df.shape[1]} колонок',
        f'\tКолонки датасета: {df.columns.tolist()}\n', sep='\n')
    
def show_item(num, name, action):
    """
        Отображает любой шаг анализа, заданный через func, в виде блока с заголовком.
    """
    print(f'{num}. {name}', '=' * 20)
    action()    
    
def show(start_num, *name_action):
    """
        Отображает шаги анализа с заголовком, нумерацией и разделением.
        Стартовая нумерация задайтся в start_num.
        Шаги переаются списком аргументов переменной длины: name, action. 
    """
    if len(name_action) % 2:
        raise ValueError('Нечётное число аргументов. В show передавайте пары значений name, action как единый список аргументов.')
    
    num_args = len(name_action)
    num = start_num
    
    for i in range(0, num_args, 2):
        if i != 0:
            print('\n')
        show_item(num, name_action[i], name_action[i + 1])
        num += 1
        if i < i - 2:
            print('\n')
            
def get_missing_values_statistics(df, name=None):
    """
        Проверяет все колонки DataFrame df на пропуски: np.nan, None, pd.NA.
        Возвращает DataFrame, содержащий две колоки: число пропусков и % пропусков по мнению функции isna( ).        
        В качестве индекса выступают имена колонок. Возвращаются только строки с ненулевыми данными.
    """
    subject = df if name is None else df[name]
    missing_count_seria = subject.isna().sum().loc[lambda v: v > 0].sort_values(ascending=False)
    missing_fraction = round(missing_count_seria / df.shape[0] * 100, 2)
    result = pd.DataFrame({'Пропущено (np.nan, None, pd.NA)': missing_count_seria, 'Доля пропусков %': missing_fraction})
    
    return 'В колонках DataFrame нет пропусков' if result.empty else result

def normalize_text_cols(df, *col_names, to_lower=True):
    """
        Конвертирует колонки col_names в string, обрезает пробелы с концов и нормализует пробелы внутри строк.
        Если задано to_lower, то приводит к нижнему регистру.
    """
    for name in col_names:        
        df[name] = df[name].astype('string').str.strip().str.replace(r'\s+', ' ', regex=True)
        if to_lower:
            df[name] = df[name].str.lower()
 
def get_whiskers(values, coeff=1.5, left=None, right=None):
    """
        Возвращает усы ящика для серии values.
    """
    Q1 = values.quantile(0.25)
    Q3 = values.quantile(0.75)
    IQR = Q3 - Q1
    
    w_left = Q1 - coeff * IQR
    w_right = Q3 + coeff * IQR
    
    if left is not None:
        w_left = max(left, w_left)
    if right is not None:
        w_right = min(right, w_right)
        
    return w_left, w_right

def get_reference_info(df, name, numeric=False):
    """
        Выполняет проверку серии name из DataFrame df как категории. А именно:
            - выводит TOP 150 уникальных значений серии и общее число уникальных значений;
            - выводит счётчик встречаемости и % для каждого значения в серии name;
            - если серия числовая, то ещё выводит min и max значения.
        Таким образом, помогает оценить качество справочника или просто серии (колонки) и заметить неявные дубликаты.
    """
    values = df[name].drop_duplicates().sort_values()
    lst = values.tolist()
    
    display(Markdown(f"**{name}:** `{lst[:150]}` = **{len(lst)}**\n\n"))
        
    counts_df = df[name].value_counts(dropna=False).astype('float64').to_frame('count')
    counts_df.index = counts_df.index.map(lambda x: '<NA>' if pd.isna(x) else x)
    counts_df['%'] = (counts_df['count'] / counts_df['count'].sum() * 100).round(2)        
    counts_df['count'] = counts_df['count'].astype('Int64')
             
    display(counts_df)
            
    print('\nПропущено:', df[name].isna().sum())
    
    if numeric:
        values_num = pd.to_numeric(df[name], errors='coerce')        
        min_val = values_num.min()
        max_val = values_num.max()
        print(f'min = {min_val}, mean = {round(values_num.mean(), 2)}, max = {max_val}, median = {values_num.median()}')
                
        whisker_left, whisker_right = get_whiskers(values_num, left=min_val, right=max_val)
        print(f'left_whisker = {whisker_left}, right_whisker = {whisker_right}')

        non_numeric = values[values_num.isna()].tolist()
        if non_numeric:
            print(f'Нецифровые значения: {non_numeric}')
    print('\n')
    
def get_frequency_report(df, name):
    """
        Возвращает частотный отчёт по колонке данных: все характеристики повторяемости значений в разных строках.
    """
    counts = df[name].value_counts(dropna=False)
    return pd.Series({
        'Среднее': counts.mean(),
        'Медиана': counts.median(), 
        'Мода': counts.mode().iloc[0] if len(counts.mode()) > 0 else None,
        'Макс': counts.max(),
        'Мин': counts.min()
    }), counts

def report_row_removing(before, after):
    print(f'✅ Из датасета в {before} строк удалено 🗑️ {before - after} строк,\n'
        f'\tосталось: {after} строк,\n'
        f'\tпроцент потерь: {round((before - after) / before * 100, 2)}%.')
````

````python
@dataclass
class BoxplotParameters:
    q1: float
    median: float
    q3: float
    iqr: float
    lower_whisker: float
    upper_whisker: float

def get_boxplot_params(df: pd.DataFrame, name: str) -> BoxplotParameters:
    # Удаляем пропуски, чтобы не ломались квантили
    series = df[name].dropna()
    
    # Расчет основных квартилей
    q1 = float(series.quantile(0.25))
    median = float(series.quantile(0.50))
    q3 = float(series.quantile(0.75))
    
    # Межквартильный размах
    iqr = q3 - q1
    
    # Теоретические границы усов
    lower_whisker_calc = q1 - 1.5 * iqr
    upper_whisker_calc = q3 + 1.5 * iqr
    
    # Реальные границы усов (обрезаем по крайним фактическим точкам)
    lower_whisker = float(series[series >= lower_whisker_calc].min())
    upper_whisker = float(series[series <= upper_whisker_calc].max())
    
    # Возвращаем заполненный дата-класс
    return BoxplotParameters(
        q1=q1,
        median=median,
        q3=q3,
        iqr=iqr,
        lower_whisker=lower_whisker,
        upper_whisker=upper_whisker
    )
````

````python
plt.figure(figsize=(18, 12))
ax = sns.countplot(data=rest_df, x='category', hue='chain')

c1_heights = [float(bar.get_height() or 0) for bar in ax.containers[0]]
c2_heights = [float(bar.get_height() or 0) for bar in ax.containers[1]]
category_sums = [h1 + h2 for h1, h2 in zip(c1_heights, c2_heights)]

for container in ax.containers:
    labels = []
    for i, bar in enumerate(container):
        height = float(bar.get_height() or 0)
        total_in_cat = category_sums[i]
        
        if height > 0 and total_in_cat > 0:
            percent = height / total_in_cat
            labels.append(f'{int(height)}\n({percent:.1%})')
        else:
            labels.append('')
    
    texts = ax.bar_label(container, labels=labels, fontsize=15, padding=8)
    for t in texts:
        t.set_rotation(50)        

ax.set_title('Сетевые и несетевые заведения по категориям', fontsize=20, pad=20)
ax.set_xlabel('Категория', fontsize=16, labelpad=20)
ax.set_ylabel('Количество точек питания', fontsize=16, labelpad=20)

ax.set_ylim(0, ax.get_ylim()[1] * 1.1)

plt.xticks(rotation=45, ha='right', fontsize=13)
plt.yticks(fontsize=13)
plt.tight_layout()
plt.show()
````

👉 Теперь посмотрим seats как count и % в разрезе category. Обратим внимание на выброс >= 1000 мест (залы) и зону Еда на вынос - нуль мест.

````python
# На основе ящика с усами оценим заведений, имеющих такие числа посадочных мест:
# 0, NA, < 5, [5, 1000) и > 1000 в разрезе category
counts = (
    rest_df.groupby('category', dropna=True)
    .agg(
        seats_na=('seats', lambda s: s.isna().sum()),
        seats_0=('seats', lambda s: (s == 0).sum()),
        seats_lt_5=('seats', lambda s: (s < 5).sum()),
        # Использован оператор & и скобки
        seats_5_1000=('seats', lambda s: ((s >= 5) & (s < 1000)).sum()),
        # Исправлено на строго больше, чтобы не дублировать 1000
        seats_gt_1000=('seats', lambda s: (s >= 1000).sum()),
    )
)

display(counts)

# И тоже самое в процентах
counts_pct = (
    rest_df.groupby('category', dropna=True)
    .agg(
        # Считаем общее количество строк в категории, включая пропуски
        total_rows=('seats', 'size'),
        
        # Считаем доли в процентах
        seats_na_pct=('seats', lambda s: (s.isna().sum() / len(s)) * 100),
        seats_0_pct=('seats', lambda s: ((s == 0).sum() / len(s)) * 100),
        seats_lt_5_pct=('seats', lambda s: ((s < 5).sum() / len(s)) * 100),
        seats_5_1000_pct=('seats', lambda s: (((s >= 5) & (s < 1000)).sum() / len(s)) * 100),
        seats_gt_1000_pct=('seats', lambda s: ((s >= 1000).sum() / len(s)) * 100),
    )
)

# Округляем до двух знаков после запятой для красоты
counts_pct = counts_pct.round(2)
display(counts_pct)
````

👉 Проблемы и особенности данных.
Очень много пропусков, лидеры 32 - 49%, особенно кафе и столовые. Пропуски скорее системные, и удалить без значимой потери данных не получится. Меньше всего пропусков в пиццериях, ресторанах и пабах 32-38%.

"Еда на вынос" распространена среди булочных и фастфуда (6% и 5%). В остальных категориях значения < 5 единичны и похожи на ошибки данных.

Доля очень крупных банкетных залов и футкортов > 1000 мест, очень мала, < 0.4%. 

👉 Из-за сильной ассиметрии и правых выбросов, для оценки seats по category, нам нужны медианы.

````python
seats_stats = (
    rest_df
    .groupby('category', dropna=True)
    .agg(
        seats_mediane=('seats', 'median'),          # медиана, NA игнорируются
        seats_mode=('seats', lambda s: s.mode().iloc[0] if not s.mode().empty else None)   # мода, если есть
    )
).sort_values(by=['seats_mediane', 'seats_mode'], ascending=True)

display(seats_stats)
````

👉 Медианы показывают, что типичная вместимость заведений колеблится 50 - 86  мест. Формат "Еда на вынос" подстверждается модой для булочных и фастфуда.
````python
def show_phink_matrix_in(df, cols, interval_cols, name, drop_names):
    pm = df[cols].phik_matrix(interval_cols=interval_cols)
    rating_corrs = pm[name].sort_values(ascending=False)
    # транспонируем, чтобы рисовалось слева на право
    plot_data = rating_corrs.to_frame().drop(drop_names).T

    plt.figure(figsize=(16, 4))
    ax = sns.heatmap(plot_data, annot=True, fmt=".2f", cmap='coolwarm', cbar=False, annot_kws={'size': 14})
    
    ax.tick_params(axis='both', which='major', labelsize=14)
    plt.ylabel('')
        
    return plt, rating_corrs
````

````python
def plot_rating_by_category(df, column, order=None, title=None, xlabel=None, figsize=(12, 6)):
    
    # Копируем данные и заполняем пропуски
    viz_df = df.copy()
    col_vis = f'{column}_vis'
    
    if viz_df[column].dtype == 'string' or viz_df[column].dtype == 'object':
        viz_df[col_vis] = viz_df[column].fillna('не указано')
    else:
        viz_df[col_vis] = viz_df[column].fillna('не указано').astype(str)
    
    # Определяем порядок категорий
    if order is None:
        # По умолчанию: сортировка по убыванию среднего рейтинга
        order = viz_df.groupby(col_vis)['rating'].mean().sort_values(ascending=False).index.tolist()
        
        # Перемещаем 'не указано' в конец, если есть
        if 'не указано' in order:
            order.remove('не указано')
            order.append('не указано')
    
    # Создаем фигуру
    plt.figure(figsize=figsize)
    
    # Основной график: линия средних значений
    ax = sns.pointplot(
        data=viz_df,
        x=col_vis,
        y='rating',
        order=order,
        errorbar=None,
        markers='o',
        linestyles='-',
        color='red',
        zorder=2
    )
    
    # Добавляем точки распределения
    sns.stripplot(
        data=viz_df,
        x=col_vis,
        y='rating',
        order=order,
        alpha=0.15,
        size=4,
        jitter=0.15,
        color='#7EB6D4',
        ax=ax,
        zorder=1
    )
    
    # Настройка отступов по Y
    ymin, ymax = viz_df['rating'].min(), viz_df['rating'].max()
    dy = (ymax - ymin) * 0.05
    ax.set_ylim(ymin - dy, ymax + dy)
    
    # Сетка и подписи
    ax.yaxis.grid(True, which='major', linestyle='--', alpha=0.5)
    ax.set_xlabel(xlabel if xlabel else column, fontsize=14)
    ax.set_ylabel('Рейтинг', fontsize=14)
    plt.title(title if title else f'Зависимость rating от {column}', fontsize=16)
    
    # Добавляем значения средних и количество записей
    for i, cat in enumerate(order):
        data_subset = viz_df[viz_df[col_vis] == cat]['rating']
        if len(data_subset) > 0:
            mean_val = data_subset.mean()
            count = len(data_subset)
            
            # Среднее значение над точкой
            ax.text(i, mean_val + 0.1, f'{mean_val:.2f}',
                    ha='center', va='bottom', fontweight='bold', 
                    fontsize=14, color='red')
            
            # Количество под осью X
            ax.text(i, ymin - dy * 0.15, f'n={count}',
                    ha='center', va='top', fontsize=12, color='gray')
    
    plt.tight_layout()
    plt.show()

def plot_rating_by_category_horizontal(df, column, order=None, title=None, xlabel=None, figsize=(12, 8)):
    
    # Копируем данные и заполняем пропуски
    viz_df = df.copy()
    col_vis = f'{column}_vis'
    
    if viz_df[column].dtype == 'string' or viz_df[column].dtype == 'object':
        viz_df[col_vis] = viz_df[column].fillna('не указано')
    else:
        viz_df[col_vis] = viz_df[column].fillna('не указано').astype(str)
    
    # Определяем порядок категорий
    if order is None:
        # По умолчанию: сортировка по убыванию среднего рейтинга
        order = viz_df.groupby(col_vis)['rating'].mean().sort_values(ascending=False).index.tolist()
        
        # Перемещаем 'не указано' в конец, если есть
        if 'не указано' in order:
            order.remove('не указано')
            order.append('не указано')
    
    # Создаем фигуру
    plt.figure(figsize=figsize)
    
    # Основной график: линия средних значений (горизонтальная)
    ax = sns.pointplot(
        data=viz_df,
        y=col_vis,  # меняем x и y местами
        x='rating',  # меняем x и y местами
        order=order,
        errorbar=None,
        markers='o',
        linestyles='-',
        color='red',
        zorder=2
    )
    
    # Добавляем точки распределения (горизонтальные)
    sns.stripplot(
        data=viz_df,
        y=col_vis,  # меняем x и y местами
        x='rating',  # меняем x и y местами
        order=order,
        alpha=0.15,
        size=4,
        jitter=0.15,
        color='#7EB6D4',
        ax=ax,
        zorder=1
    )
    
    ax.tick_params(axis='y', pad=20)
    
    # Настройка отступов по X
    xmin, xmax = viz_df['rating'].min(), viz_df['rating'].max()
    dx = (xmax - xmin) * 0.05
    ax.set_xlim(xmin - dx, xmax + dx)
    
    # Сетка и подписи
    ax.xaxis.grid(True, which='major', linestyle='--', alpha=0.5)
    ax.set_ylabel(xlabel if xlabel else column, fontsize=14)
    ax.set_xlabel('Рейтинг', fontsize=14)
    plt.title(title if title else f'Зависимость rating от {column}', fontsize=16)
    
    # Добавляем значения средних (без счётчиков)
    for i, cat in enumerate(order):
        data_subset = viz_df[viz_df[col_vis] == cat]['rating']
        if len(data_subset) > 0:
            mean_val = data_subset.mean()
            
            # Среднее значение справа от точки
            ax.text(mean_val + 0.05, i, f'{mean_val:.2f}',
                    ha='left', va='center', fontweight='bold', 
                    fontsize=14, color='red')
    
    plt.tight_layout()
    plt.show()
    

price_order = ['низкие', 'средние', 'выше среднего', 'высокие', 'не указано']

plot_rating_by_category(
    rest_df,
    'price',
    price_order,
    'Зависимость rating от ценовой категории (price)',
    'Уровень цен'
)

plot_rating_by_category_horizontal(
    rest_df,
    'district',
    title='Зависимость рейтинга от района (district)',
    xlabel='Район',
    figsize=(12, 10)  # чуть больше высоты для длинных названий
)

plot_rating_by_category(
    rest_df,
    'category',    
    title='Зависимость рейтинга от категории заведения',
    xlabel='Категория'    
)
````

Выяснилось, что наиболее сильную связь с рейтингом показывают ценовая категория (0.34) и район расположения (0.32). Чуть слабее связана категория заведения (0.27).     Медианный чек по Москве — **750 ₽**. Влияние цен на рейтинг прямое: при переходе от низкого ценового сегмента к сегменту «выше среднего» средний рейтинг вырастает **с 4.17 до 4.44 баллов**.

    Анализ корреляции (Phi_K) подтвердил: сильнее всего на рейтинг влияют **price** (коэф. **0.35**) и **локация/район** (коэф. **0.32**). Сетевой статус и размер зала оказались второстепенными факторами.

### 5.2 Рекомендации для заказчика

1.  **Локационная стратегия**  
    Приоритет — **центр (ЦАО)**, где сконцентрировано **27% рынка** и самые высокие чеки. В спальных районах (ЮВАО, СВАО) лучше открывать доступные форматы «у дома» со средним чеком **500–650 ₽**, так как попытки запустить там дорогие проекты ведут к падению среднего рейтинга на **0.2–0.3 балла**.

2.  **Выбор категорий и целевой рейтинг**  
    **Перспективные ниши:** пабы (рейтинг **4.39**) и пиццерии (стабильно высокие оценки). Ориентир для нового бизнеса — планка рейтинга **не ниже 4.2–4.3**, так как основной пул конкурентов (включая лидеров вроде «Додо Пиццы» с рейтингом **4.29**) находится именно в этом диапазоне.

3.  **Ценовой сегмент и позиционирование**  
    Оптимальная стратегия — **«выше среднего»**. Этот сегмент дает наивысший средний рейтинг (**4.44**) при адекватном уровне рисков. Слишком низкий чек грозит рейтингом **ниже 4.17**, а премиум требует исключительной локации и сервиса.

4.  **Масштабирование и сеть**  
    Для франчайзинга идеальны **кофейни** и **булочные**. Рекомендуется изучить опыт сети **«Буханка»**: при низком среднем чеке **~212 ₽** они удерживают выдающийся рейтинг **4.42**, что говорит о максимальной лояльности аудитории к формату «пекарня у дома».

5.  **Формат зала**  
    Не инвестируйте в огромные залы. Типичный размер зала (медиана) варьируется от 50 мест (булочные) до 86 мест (рестораны). Кофейни в среднем имеют 80 мест, бары — 82 места, пиццерии — 55 мест. Размер зала практически не коррелирует с рейтингом, поэтому излишки площади лучше направить на создание уюта или летнюю веранду.

6.  **Круглосуточный режим**  
    Круглосуточная работа (**24/7**) сама по себе **не повышает рейтинг** (слабая отрицательная корреляция). Вводить такой режим стоит только при гарантированном ночном трафике, иначе это приведет лишь к росту издержек.

7.  **Дальнейшая работа с данными**  
    Улучшить сбор ценовых данных. Сейчас **48%** информации о чеках либо отсутствует, либо получена через некорректный парсинг. Для точной юнит-экономики необходимо восстановить данные хотя бы до уровня **80–85%** заполняемости.
