````python
def show_item(num, name, action):
    """
        Отображает любой шаг анализа, заданный через func, в виде блока с заголовком.
    """
    print(f'{num}. {name}', '=' * 20)
    action()    
    
def show(start_num, *name_action):
    """
        Отображает шаги анализа с заголовком, нумерацией и разделением.
        Стартовая нумерация задайтся в start_num.
        Шаги переаются списком аргументов переменной длины: name, action. 
    """
    if len(name_action) % 2:
        raise ValueError('Нечётное число аргументов. В show передавайте пары значений name, action как единый список аргументов.')
    
    num_args = len(name_action)
    num = start_num
    
    for i in range(0, num_args, 2):
        if i != 0:
            print('\n')
        show_item(num, name_action[i], name_action[i + 1])
        num += 1
        if i < i - 2:
            print('\n')
    
def get_missing_values_statistics(df, name=None):
    """
        Проверяет все колонки DataFrame df на пропуски: np.nan, None, pd.NA.
        Возвращает DataFrame, содержащий две колоки: число пропусков и % пропусков по мнению функции isna( ).        
        В качестве индекса выступают имена колонок. Возвращаются только строки с ненулевыми данными.
    """
    subject = df if name is None else df[name]
    missing_count_seria = subject.isna().sum().loc[lambda v: v > 0].sort_values(ascending=False)
    missing_fraction = round(missing_count_seria / df.shape[0] * 100, 2)
    result = pd.DataFrame({'Пропущено (np.nan, None, pd.NA)': missing_count_seria, 'Доля пропусков %': missing_fraction})
    
    return 'В колонках DataFrame нет пропусков' if result.empty else result

def get_reference_info(df, name, numeric=False):
    """
        Выполняет проверку серии name из DataFrame df относительно роли справочника. А именно:
            - выводит TOP 150 уникальных значений серии и общее число уникальных значений;
            - выводит счётчик встречаемости для каждого значения в серии name;
            - если серия числовая, то ещё выводит min и max значения.
        Таким образом, помогает оценить качество справочника или просто серии (колонки) и заметить неявные дубликаты.
    """
    values = df[name].drop_duplicates().sort_values()
    lst = values.tolist()
    print(f'{name}: ', lst[:150], len(lst), end='\n\n')
    print(df[name].value_counts(dropna=False), end='\n')
    if numeric:
        values_num = pd.to_numeric(values, errors='coerce')
        non_numeric = values[values_num.isna()].tolist()
        print(f'min = {values_num.min()}, max = {values_num.max()}')
        if non_numeric:
            print(f'Нецифровые значения: {non_numeric}')
    print('\n')
    
def get_frequency_report(df, name):
    """
        Возвращает частотный отчёт по колонке данных: все характеристики повторяемости значений в разных строках.
    """
    counts = df[name].value_counts(dropna=False)
    return pd.Series({
        'Среднее': counts.mean(),
        'Медиана': counts.median(), 
        'Мода': counts.mode().iloc[0] if len(counts.mode()) > 0 else None,
        'Макс': counts.max(),
        'Мин': counts.min()
    }), counts
````

 #### Сухой остаток после анализа пропусков и уникальности значений в колонках

Мы загружали CSV через pd.read_csv с настройками по-умолчанию. Pandos в этом случае использует в зависимости от версии свои типы или NumPy, и иногда откатывается в типу Python object, который может в одной серии хранить разные типы. Это не очень эффективно по потреблению памяти. Массивы NumPy и array могут хранить только один скалярный тип данных. Потому и надо заменять все пропуски на 

Бывают три вида пропусков:
 - пусто ",,";
 - стандартный маркер: NA, NaN, null;
 - кастомный маркер: -, #, tbd, none, unknown и тд.
 
 Если в числовой колонке есть только первые 2 вида, то будет числовая колонка с np.nan или pd.NA, в зависимости от версии. При этом, она откатится от int64 к float64, в общем виде. 
 Если в числовой колонке есть кастомные маркеры пропусков, то серия откатится к object и будет хранить числа и строки.
 С булевыми типами аналогично.
 Надо просмотреть колонку и найти кастомные маркеры пропусков, чтобы их указать Pandos и не было отката. Тут это и проделаем на основе анализа выше.
 
| Column          | Dtype   | Comment |
| --------------- | ------- | ------- |
| Name            | object  | строка даёт object, есть пропуски 2шт     |
| Platform        | object  | строка даёт object, нет пропусков        |
| Year of Release | float64 | Есть 275 пропусков, поэтому, откат в float и использование np.nan        |
| Genre           | object  | Тут строка и 2 пропуска        |
| NA sales        | float64 | Исходно float без пропусков, поэтому float64        |
| EU sales        | object  | Есть строковый маркер 'unknown', поэтому откат к object: смешенная серия       |
| JP sales        | object  | Есть строковый маркер 'unknown', поэтому откат к object: смешенная серия        |
| Other sales     | float64 | Исходно float без пропусков, поэтому float64        |
| Critic Score    | float64 | Исходно float с пропусками, поэтому float64        |
| User Score      | object  | Исходно float с пропусками и маркером 'tbd', откат к object        |
| Rating          | object  | строка даёт object        |
 
Таким образом, нужно задать конвертацию с оптимальными типами и обработку кастомных пропусков tbd и unknown, чтобы они превратились в pd.NA. 

````python
# теперь подготовим чистовой DataFrame: справильныи типами колонок, именами и обработанными пропусками данных

all_cols = {'year_of_release': 'Int64', 
              'critic_score': 'Int64', 
              'na_sales': 'Float64', 
              'eu_sales': 'Float64', 
              'jp_sales': 'Float64', 
              'other_sales': 'Float64',
              'user_score': 'Float64',
              'name': 'string',
              'platform': 'string',
              'genre': 'string',
              'rating': 'string',
             }
numeric_cols = [col_name for col_name, data_type in all_cols.items() if data_type in ('Int64', 'Float64')]

# df = pd.read_csv(DATASET_PATH, dtype=data_types, na_values=['tbd', 'unknown'], skipinitialspace=True).rename(columns=lambda col_name: col_name.replace(' ', '_').lower())

# применяем правила Snake case ко всем колонкам DataFrame
df = df.rename(columns=lambda col_name: col_name.replace(' ', '_').lower())

df = df.replace(['tbd', 'unknown'], pd.NA)
df[numeric_cols] = df[numeric_cols].apply(pd.to_numeric, errors='coerce')
df = df.astype(all_cols)
````
