 ### Описание данных

- **Name** — название игры.  
- **Platform** — название платформы.  
- **Year of Release** — год выпуска игры.  
- **Genre** — жанр игры.  
- **NA sales** — продажи в Северной Америке (в миллионах проданных копий).  
- **EU sales** — продажи в Европе (в миллионах проданных копий).  
- **JP sales** — продажи в Японии (в миллионах проданных копий).  
- **Other sales** — продажи в других странах (в миллионах проданных копий).  
- **Critic Score** — оценка критиков (от 0 до 100).  
- **User Score** — оценка пользователей (от 0 до 10).  
- **Rating** — рейтинг организации ESRB (англ. *Entertainment Software Rating Board*), которая присваивает играм возрастные категории.

Справочник Rating:
| Рейтинг | Описание | Возрастная категория | Период действия |
| :--- | :--- | :--- | :--- |
| **RP** | **Rating Pending** — ожидается окончательный рейтинг. | Любой | 1994 — н.в. |
| **EC** | **Early Childhood** — для дошкольников. Без неприемлемого контента. | 3+ | 1994 — 2018 |
| **E** | **Everyone** — для всех (до 1998 г. назывался **K-A**). | 6+ | 1994 — н.в. |
| **E10+** | **Everyone 10+** — умеренное насилие и грубая речь. | 10+ | 2005 — н.в. |
| **T** | **Teen** — для подростков. Допустимы кровь и сексуальный подтекст. | 13+ | 1994 — н.в. |
| **M** | **Mature** — для лиц от 17 лет. Реалистичное насилие и нагота. | 17+ | 1994 — н.в. |
| **AO** | **Adults Only** — строго для взрослых. Высокий уровень воздействия. | 18+ | 1994 — н.в. |

#### Подробные описания:

*   **RP (Rating Pending)** — используется в рекламных материалах для игр, которым ещё не присвоен окончательный рейтинг ESRB. Принят в 1994 году и действует до сих пор.
*   **EC (Early Childhood)** — игры, ориентированные на дошкольную аудиторию. Не содержат материалов, которые родители могли бы считать неприемлемыми. Принят в 1994 году и действовал до 2018 года.
*   **E (Everyone)** — игры для всех возрастов. Могут содержать незначительное количество насилия и сленга. До 1998 года рейтинг был известен как Kids to Adults (K-A). Принят в 1994 году и действует до сих пор.
*   **E10+ (Everyone 10 and older)** — игры для людей от 10 лет и старше. Могут содержать более высокий уровень насилия, грубую разговорную речь. Принят в 2005 году и действует до сих пор.
*   **T (Teen)** — игры для людей от 13 лет и старше. Могут содержать умеренное количество насилия, включая небольшие количества крови, грубую разговорную речь и юмор, сексуальное содержание. Принят в 1994 году и действует до сих пор.
*   **M (Mature)** — игры для лиц от 17 лет и старше. Могут содержать более интенсивные или реалистичные изображения насилия, включая кровь, ужасы, увечья и изображения смерти, непристойные сексуальные сцены, частичную наготу и грубую ненормативную лексику. Принят в 1994 году и действует до сих пор.
*   **AO (Adults Only)** — игры для людей от 18 лет и старше. Содержат материалы с более высоким уровнем воздействия, чем рейтинг M. Принят в 1994 году и действует до сих пор.

```python
def show(num, name, func):
    print('\n', f'{num}. {name}', '-' * 20, '\n')
    func()

def get_missings(df):
    missing = df.isna().sum().loc[lambda v: v > 0].sort_values(ascending=False)
    missing_fraction = round(missing / df.shape[0] * 100, 2)
    return pd.DataFrame({'N/A': missing, 'Fraction %': missing_fraction})

def get_reference_info(df, name, numeric=False):
    values = df[name].drop_duplicates().sort_values()
    lst = values.tolist()
    print(f'{name}: ', lst[:150], len(lst), end='\n\n')
    print(df[name].value_counts(dropna=False), end='\n\n')
    if numeric:
        values_num = pd.to_numeric(values, errors='coerce')
        print(f'min = {values_num.min()}, max = {values_num.max()}\n\n')
```

```python
df = pd.read_csv('./datasets/new_games.csv') # черновая загрузка датасета

# ознакомление со структурой

print(f'Всего: {df.shape[0]} строк, {df.shape[1]} колонок')
print(f'Колонки: {df.columns.tolist()}')

show(1, 'info', lambda: df.info())
show(2, 'missing', lambda: display(get_missings(df)))
show(3, 'describe', lambda: print(df.describe()))
show(4, 'display', lambda: display(df))
```

```python
# пропуски уже помечены типом pd.NA при Чистовой загрузке CSV, это лучше значений-индикаторов
```

```python
# заменяем пропущенные Sales на средние значения в группе platform, year_of_release

eu_means = df.groupby(['platform', 'year_of_release'])['eu_sales'].transform('mean')
jp_means = df.groupby(['platform', 'year_of_release'])['jp_sales'].transform('mean')
df['eu_sales'] = df['eu_sales'].fillna(eu_means)
df['jp_sales'] = df['jp_sales'].fillna(jp_means)

# это для пустых групп, чтобы не появилось N/A
df[['eu_sales', 'jp_sales']] = df[['eu_sales', 'jp_sales']].fillna(0)

display(df.loc[empty_sales_rows.index]) # теперь все Sales pfgjkytys
```

## Категоризация данных
    
Проведите категоризацию данных:
- Разделите все игры по оценкам пользователей и выделите такие категории: высокая оценка (от 8 до 10 включительно), средняя оценка (от 3 до 8, не включая правую границу интервала) и низкая оценка (от 0 до 3, не включая правую границу интервала).

```python
df_actual['user_score_cat'] = pd.cut(df_actual['user_score'], bins=[0, 3, 8, 11], labels=["Низкая", "Средняя", "Высокая"], right=False)
df_actual['user_score_cat'] = df_actual['user_score_cat'].astype('string') # для консистентности
```

- Разделите все игры по оценкам критиков и выделите такие категории: высокая оценка (от 80 до 100 включительно), средняя оценка (от 30 до 80, не включая правую границу интервала) и низкая оценка (от 0 до 30, не включая правую границу интервала).

```python
df_actual['critic_score_cat'] = pd.cut(df_actual['critic_score'], bins=[0, 30, 80, 101], labels=["Низкая", "Средняя", "Высокая"], right=False)
df_actual['critic_score_cat'] = df_actual['critic_score_cat'].astype('string') # для консистентности
```

```python
print(df_actual.groupby('user_score_cat', dropna=False)['name'].count(), '\n')
print(df_actual.groupby('critic_score_cat', dropna=False)['name'].count())
```
